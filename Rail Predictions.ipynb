```python
# Future prediction analysis based on the historic data from previous years 
# Mode of Transport - Rail


# Import the required libraries
import pandas as pd     #Pandas for data cleaning, manipulation and time series
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the data into a Pandas DataFrame
data = pd.read_csv('Transport_Rail.csv')

# Pivot to have modes of transport as columns and weeks of the year as rows
data = data.pivot( columns='Mode of Transport', values='VALUE')

# Update the index to include the weeks of 2024
data = data.reindex(list(data.index) + list(range(53, 105)), fill_value=None)

# Create separate DataFrames for each mode of transport
#Here dataframe created for rail mode of transport
rail_data = data[['Rail']]

# Drop rows with missing values
data.dropna(inplace=True)
# Drop rows with missing values
rail_data.dropna(inplace=True)

# Create a linear regression model for each mode of transport
rail_model = LinearRegression()

# Fit the models using the data
rail_model.fit(rail_data.index.values.reshape(-1, 1), rail_data.values.reshape(-1, 1))

# Predict the number of passengers for the next 52 weeks, which includes weeks of 2024
rail_predictions = rail_model.predict([[i] for i in range(1, 53+52)])

# Print the predictions for weeks of 2024
print("Rail predictions for 2024:")
for i in range(52, len(rail_predictions)):
    print(f"Week {i-51}: {int(round(rail_predictions[i][0]))}")
    


# Plot the actual rail passenger data
plt.plot(data.index, data['Rail'], label='Actual')

# Plot the predicted rail passenger data for 2024
plt.plot(range(1, 53+52), rail_predictions, label='Predicted')

# Add labels and legend
plt.xlabel('Weeks')
plt.ylabel('Number of passengers')
plt.title('Passenger data and future predictions for 2024 for Rail mode of transport ')
plt.legend()

# Show the plot
plt.show()
# ___________________________________________________________________________________________________
```
